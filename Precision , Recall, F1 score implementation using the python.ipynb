```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_data=pd.read_csv('/content/heart_disease_data.csv')
```

```python
# 0 -> healthy heart
#1 -> defective heart

x=heart_data.drop(columns='target', axis=1)
y=heart_data['target']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape, x_test.shape,x_train.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
y_train_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print ('accuracy of training data:', y_train_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
y_test_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print ('accuracy of training data:', y_test_accuracy)
```

```python
#precision

from sklearn.metrics import  precision_score
```

```python
# precision for the training data
precision_train=precision_score(y_train,x_train_prediction)
print('Traning data precision=' , precision_train)
```

```python
# precision for the testing data
precision_test=precision_score(y_test,x_test_prediction)
print('Testing data precision=' , precision_test)
```

```python

# Recall

from sklearn.metrics import recall_score
```

```python
# Recall for the testing data
Recall_train=recall_score(y_train,x_train_prediction)
print('Traning data Recall=' , Recall_train)
```

```python

# precision for the testing data
Recall_test=recall_score(y_test,x_test_prediction)
print('Testing data Recall=' , Recall_test)
```

```python
# F 1 score
from sklearn.metrics import f1_score
```

```python
# F1 score for the testing data
f1_train=f1_score(y_train,x_train_prediction)
print('Traning data f1score=' ,f1_train)
```

```python

# F1 score for the testing data
f1_test=f1_score(y_test,x_test_prediction)
print('Testing data f1score=' , f1_test)
```

```python

def precision_recall_f1_score(true_labels, pred_labels):

  precision_value = precision_score(true_labels, pred_labels)
  recall_value = recall_score(true_labels, pred_labels)
  f1_score_value = f1_score(true_labels, pred_labels)

  print('Precision =',precision_value)
  print('Recall =',recall_value)
  print('F1 Score =',f1_score_value)
```

```python

precision_recall_f1_score(y_train,x_train_prediction)
```

```python
precision_recall_f1_score(y_test,x_test_prediction)
```
